```python
import pandas as pd
import seaborn as sns
import matplotlib

sns.set(color_codes=True)
```

## We begin by importing pandas and seaborn; two tools used to analyse and manipulate our data

The sns.set merely is added to offer the better looking plots seen below.

```python
bestsellers = pd.read_csv('bestsellers with categories.csv')
```

## Intialize the read of the csv file; we are working with the Amazon Bestselers

```python
bestsellers.head()
```

## Open the bestsellers as a head to allow for simple access to our data both for viewing and editing

```python
bestsellers.info()
```

## Info section opened to view usage as well as Columns to be manipulated

```python
sns.barplot(bestsellers['User Rating'], bestsellers['Reviews'])
```

## Simple bargraph is created to show the distribution between Reviews and Ratings

```python
sns.stripplot(bestsellers['User Rating'], bestsellers['Genre'], jitter = True)
```

## Stripplot highlighting the Ratings vs Genre 

```python
sns.jointplot(bestsellers['Year'], bestsellers['User Rating'])
```

## Joinplot comparing the Ratings to the year each Bestseller was released

```python
sns.distplot(bestsellers['Price'])
```

## Distributed Plot for prevalence of price in selected area; 'Bell' Curve around ~$10 

```python
sns.jointplot(bestsellers['User Rating'], bestsellers['Price'], kind="hex")
```
